```python
# ==========================================
# Course Registration System
# ==========================================
import getpass

# ==========================================
# Users Database
# ==========================================
users = {}

# ==========================================
# Teachers Database
# Each teacher is assigned to one subject
# ==========================================
teachers = {
    "math_teacher": {
        "password": "1111",
        "subject": "Math"
    },

    "physics_teacher": {
        "password": "2222",
        "subject": "Physics"
    },

    "programming_teacher": {
        "password": "3333",
        "subject": "Programming"
    },

    "networks_teacher": {
        "password": "4444",
        "subject": "Networks"
    },

    "ai_teacher": {
        "password": "5555",
        "subject": "AI"
    }
}

# ==========================================
# Courses Database
# ==========================================
courses = {
    "Math": {
        "students": []
    },

    "Physics": {
        "students": []
    },

    "Programming": {
        "students": []
    },

    "Networks": {
        "students": []
    },

    "AI": {
        "students": []
    }
}

# ==========================================
# Student Registration
# ==========================================
def register():

    print("\n===== STUDENT REGISTRATION =====")
    username = input("Create Username: ")
    password = getpass.getpass("Create Password: ")

    if username == "" or password == "":
        print("Fields cannot be empty.")
        return

    if username in users:
        print("Username already exists.")
        return

    print("\n===== AVAILABLE SUBJECTS =====")

    for subject in courses.keys():
        print("-", subject)

    print("\nType subject names separated by commas")
    print("Example: Math, AI, Physics")

    selected = input("\nEnter Subjects: ")
    selected_courses = []

    try:
        choices = selected.split(",")

        for choice in choices:
            typed_subject = choice.strip().lower()
            matched_subject = None

            for c in courses.keys():
                if c.lower() == typed_subject:
                    matched_subject = c
                    break

            if matched_subject:
                if matched_subject not in selected_courses:
                    selected_courses.append(matched_subject)

            else:
                print(f"Invalid subject: {choice.strip()}")
                return

    except:
        print("Invalid input format.")
        return

    if len(selected_courses) == 0:
        print("No subjects selected.")
        return

    users[username] = {
        "password": password,
        "courses": selected_courses
    }

    for subject in selected_courses:
        courses[subject]["students"].append(username)

    print("\nRegistration Successful!")
    print("You are registered in:")

    for subject in selected_courses:
        print("-", subject)

# ==========================================
# Student Login
# ==========================================
def student_login():

    print("\n===== STUDENT LOGIN =====")
    username = input("Enter Username: ")
    password = getpass.getpass("Enter Password: ")

    # Authentication
    if username in users and users[username]["password"] == password:
        print("Login Successful!")
        return username

    else:
        print("Invalid Username or Password.")
        return None


# ==========================================
# Teacher Login
# ==========================================
def teacher_login():

    print("\n===== TEACHER LOGIN =====")
    username = input("Enter Teacher Username: ")

    # Hidden password
    password = getpass.getpass("Enter Password: ")

    # Authentication
    if username in teachers and teachers[username]["password"] == password:
        print("Teacher Login Successful!")
        return username

    else:
        print("Invalid Teacher Username or Password.")
        return None

# ==========================================
# Forgot Password
# ==========================================
def forgot_password():

    print("\n===== PASSWORD RECOVERY =====")
    username = input("Enter Username: ")

    if username not in users:
        print("Username not found.")
        return

    answer = input("What is your favorite color? ")

    if answer == users[username]["recovery"]:
        new_password = getpass.getpass("Enter New Password: ")
        users[username]["password"] = new_password
        print("Password Reset Successful!")

    else:
        print("Incorrect Recovery Answer.")

# ==========================================
# Display Courses
# ==========================================
def display_courses():

    print("\n===== AVAILABLE COURSES =====")
    number = 1

    for course in courses:
        print(f"{number}. {course}")
        number += 1

# ==========================================
# Add Course
# ==========================================
def add_course(current_user):

    display_courses()
    choice = input("\nChoose Course Number: ")
    course_list = list(courses.keys())

    # Validation
    if not choice.isdigit():
        print("Please enter a valid number.")
        return

    choice = int(choice)

    if choice < 1 or choice > len(course_list):
        print("Invalid Course Choice.")
        return

    selected_course = course_list[choice - 1]

    # Check duplicate
    if selected_course in users[current_user]["courses"]:
        print("You are already enrolled in this course.")
        return

    # Add course to student
    users[current_user]["courses"].append(selected_course)

    # Add student to course
    courses[selected_course]["students"].append(current_user)
    print(f"You successfully enrolled in {selected_course}!")

# ==========================================
# View Student Courses
# ==========================================
def view_courses(current_user):

    print("\n===== MY COURSES =====")
    student_courses = users[current_user]["courses"]

    if len(student_courses) == 0:
        print("No Courses Enrolled.")

    else:
        for course in student_courses:
            print("-", course)

# ==========================================
# Teacher View
# Each teacher only sees their subject
# ==========================================
def teacher_view(teacher_username):

    teacher_subject = teachers[teacher_username]["subject"]
    print(f"\n===== {teacher_subject} COURSE REPORT =====")
    student_list = courses[teacher_subject]["students"]
    print(f"Total Students: {len(student_list)}")

    if len(student_list) == 0:
        print("No students enrolled.")

    else:
        print("\nRegistered Students:")

        for i, student in enumerate(student_list, start=1):
            print(f"{i}. Student Name: {student}")

# ==========================================
# Student Menu
# ==========================================
def student_menu(current_user):

    while True:
        print(f"\n===== WELCOME {current_user} =====")
        print("1. Add Course")
        print("2. View My Courses")
        print("3. Logout")
        choice = input("Enter Your Choice: ")

        if choice == "1":
            add_course(current_user)

        elif choice == "2":
            view_courses(current_user)

        elif choice == "3":
            print("Logging Out...")
            break

        else:
            print("Invalid Choice.")

# ==========================================
# Teacher Menu
# ==========================================
def teacher_menu(current_teacher):

    while True:

        print(f"\n===== TEACHER PANEL =====")
        print("1. View My Students")
        print("2. Logout")
        choice = input("Enter Your Choice: ")

        if choice == "1":
            teacher_view(current_teacher)

        elif choice == "2":
            print("Logging Out...")
            break

        else:
            print("Invalid Choice.")

# ==========================================
# Main System
# ==========================================
while True:
    print("\n===== COURSE REGISTRATION SYSTEM =====")
    print("1. Student Register")
    print("2. Student Login")
    print("3. Teacher Login")
    print("4. Forgot Password")
    print("5. Exit")
    option = input("Enter Your Choice: ")

    # Student Registration
    if option == "1":
        register()

    # Student Login
    elif option == "2":
        user = student_login()

        if user:
            student_menu(user)

    # Teacher Login
    elif option == "3":
        teacher = teacher_login()

        if teacher:
            teacher_menu(teacher)

    # Forgot Password
    elif option == "4":
        forgot_password()

    # Exit
    elif option == "5":
        print("Exiting System...")
        break

    else:
        print("Invalid Choice.")
```
